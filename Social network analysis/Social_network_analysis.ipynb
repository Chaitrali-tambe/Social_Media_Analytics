# **11.Betweeness Centrality**

```python
pos = nx.spring_layout(G)
betCent = nx.betweenness_centrality(G, normalized=True, endpoints=True)
node_color = [20000.0* G.degree(v)for v in G]
node_size = [v*10000 for v in betCent.values()]
plt.figure(figsize=(20,20))
nx.draw_networkx(G, pos=pos, with_labels = False, node_color = node_color, node_size= node_size)
sorted(betCent, key=betCent.get,reverse= True)[:5]
```

# 14.Bridges

```python
nx.has_bridges(G)
```

```python
bridges = list(nx.bridges(G))
len(bridges)
```

```python
local_bridges = list(nx.local_bridges(G, with_span = False))
len(local_bridges)
```

showing bridges and local_bridges in a network The bridges can be seen with the blue color and local_bridges with the green color. Black edges are niethre local bridges nor bridges.

```python
plt.figure(figsize=(15,20))
nx.draw_networkx(G, pos=pos, with_labels = False, node_size= node_size, width = 0.15)
nx.draw_networkx_edges(G, pos, edgelist= local_bridges, edge_color = "green", 
                        node_size= node_size, width = 0.5)
nx.draw_networkx_edges(G, pos, edgelist= bridges, edge_color = "blue", 
                        node_size= node_size, width = 0.5)
plt.axis("off")
```
